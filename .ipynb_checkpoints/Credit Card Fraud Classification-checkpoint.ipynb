## PCA

```python
from sklearn.decomposition import PCA

X = df.drop('Class', axis = 1)
y = df['Class']

pca = PCA(n_components = 2)
X_pca = pca.fit_transform(X)

# create a new dataframe to visualize
pca_df = pd.DataFrame(X_pca, columns = ['PC1', 'PC2'])
pca_df['Class'] = y

# Create scatter plot of PCA components
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PC1', y='PC2', hue='Class', data=pca_df, palette='Set1')
plt.title('PCA Visualization of Credit Card Fraud Detection Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Class')
plt.show()
```

# Model Creation and Evaluation
Models implemented: 
* Logistic Regression and 
* Random Forest Classifier
* Support Vector Machine

***Justification:***

Evaluation metric: The metrics considered as candidates include precision, recall, and accuracy.
However, the most appropruate in this case is deemed to be recall. This is because it is more useful that the model identifies(predicts correctly) which observations are actually fraudulent. Recall rate describes the proportion of actual fraudulent cases that the model is able to correctly identify. 

That said, I think that a look at the precision rate also is worth it since it may give us a fair idea of the how many non-fraudulent observations may be classified by the model as "fraudulent". Knowing this will add more context to how much reliable the model is and also to help noot to misclassify non-fraudulent transactions as fraudulent ones.


```python
### Logistic Regression
* logistic Regression - ----> || Accuracy: 0.9421 || ----> ||Precision:0.9789||
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, f1_score, recall_score
from sklearn.model_selection import train_test_split
```

```python
# split the data into training and validation set
y = df['Class']
X = df.drop('Class', axis = 1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 42)
```

```python

# fit the model
logistic_model = LogisticRegression(penalty = 'l1', solver = 'liblinear')
logistic_model.fit(X_train, y_train)

# Predict the test set on the model
y_pred = logistic_model.predict(X_test)

# Evaluate the model performance
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
print('Accuracy:', accuracy)
print('Precision:', round(precision*100, 2), '%')
print('Recall:', round(recall * 100, 2), '%')

```

```python
# Cross Validation
from sklearn.model_selection import cross_val_score
cv_scores_logistic = cross_val_score(logistic_model, X_train, y_train, cv = 5, scoring = 'recall')
print("Cross-Validation Scores:", cv_scores_logistic)
print("Average Recall:", cv_scores_logistic.mean())
```

### Random Forest Classifier

We get a model accuracy of 92.63%. Then we get.a recall score of 100%. This means of the transactions that were actually fraudulent, the model predicts with 100% accuracy. 

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.model_selection import train_test_split,cross_val_predict
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
#fit the model
rf = RandomForestClassifier(n_estimators=100, max_depth = 1)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)

# Evaluate the model on training and test
rf_train = round(rf.score(X_train, y_train) * 100, 2) # training score
rf_accuracy = round(accuracy_score(y_pred_rf, y_test) * 100, 2) # training accuracy
rf_recall = round(recall_score(y_pred_rf, y_test) * 100, 2) # recall
print("Training Accuracy :",rf_train ,"%")
print("Model Accuracy Score :",rf_accuracy ,"%")
rf_recall = print('Recall: ', rf_recall, '%')

# claissification report
print("\033[1m--------------------------------------------------------\033[0m")
print("Classification_Report: \n",classification_report(y_test,y_pred_rf))
print("\033[1m--------------------------------------------------------\033[0m")
```

```python
from sklearn.model_selection import cross_val_score 
#scoring = {'accuracy': 'accuracy', 'precision': 'precision', 'recall': 'recall'}
cv_scores = cross_val_score(rf, X_train, y_train, cv=5, scoring='recall')
print("Cross-Validation Scores:", cv_scores)
print("Average Recall:", cv_scores.mean())
```
