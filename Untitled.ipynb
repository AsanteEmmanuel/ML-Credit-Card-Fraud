```python
# get the fraud and non-fraud records
normal = scaled_X[scaled_X['Class'] == 0]
fraud = scaled_X[scaled_X['Class'] == 1]

# get samples from the majority class # downsampling in this case 50-50 classes
normal_sample = normal.sample(n = 473)
df = pd.concat([normal_sample, fraud], axis = 0)
df
```

# Data Visualisation
 Create a correlation heatmap to highlight relationships between features

It seems that there may be some informative relationship between class vs {V14, V17, V4}. also it may be worth exploring the pairwise relationships between members of the set {V9, V10} and {V16, V17, V18}
creating a boxplot to see whether v14 or Amount has a distinct effect on the transactioons nature i.e. fraud or normal

```python
import matplotlib.pyplot as plt
import seaborn as sns

#
selected_features = ['V4', 'V17']

# Create box plots for selected features
plt.figure(figsize=(12, 6))
for feature in selected_features:
    plt.subplot(1, len(selected_features), selected_features.index(feature) + 1)
    sns.boxplot(x='Class', y=feature, data=df)
    plt.title(f'Box Plot of {feature}')
    plt.xlabel('Class (0: Non-Fraud, 1: Fraud)')
    plt.ylabel(feature)
    
plt.tight_layout()
plt.show()
```

```python
#
selected_features = ['Amount', 'V14']

# Create box plots for selected features
plt.figure(figsize=(12, 6))
for feature in selected_features:
    plt.subplot(1, len(selected_features), selected_features.index(feature) + 1)
    sns.boxplot(x='Class', y=feature, data=df)
    plt.title(f'Box Plot of {feature}')
    plt.xlabel('Class (0: Non-Fraud, 1: Fraud)')
    plt.ylabel(feature)
    
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='V16', y='V17', hue='Class', data=df)
plt.title('Scatter Plot of V16 vs V17 with Class Hue')
plt.xlabel('V16')
plt.ylabel('V17')
plt.legend(title='Class', loc='upper right')
plt.show()
```

Because of the correlations between the some of the features, it is worth exploring PCA.

## PCA

```python
from sklearn.decomposition import PCA

X = df.drop('Class', axis = 1)
y = df['Class']

pca = PCA(n_components = 2)
X_pca = pca.fit_transform(X)

# create a new dataframe to visualize
pca_df = pd.DataFrame(X_pca, columns = ['PC1', 'PC2'])
pca_df['Class'] = y

# Create scatter plot of PCA components
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PC1', y='PC2', hue='Class', data=pca_df, palette='Set1')
plt.title('PCA Visualization of Credit Card Fraud Detection Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Class')
plt.show()
```
